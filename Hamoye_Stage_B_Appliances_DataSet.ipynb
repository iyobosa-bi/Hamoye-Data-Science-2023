```python
##Importing all required libraries
import numpy as np
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
#from xgboost import XGBRegressor
#from sklearn.tree import DecisionTreeRegressor
#from sklearn.neighbors import KNeighborsRegressor
from sklearn.linear_model import Ridge
#from sklearn.svm import SVR
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import MinMaxScaler

from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
```

```python
## importing Dataset from local Machine

appl_Data= pd.read_csv(r"C:\Users\Blessing_Iyobosa\Downloads\energydata_complete.csv")
appl_Data.head()
```

```python
appl_Data.describe()

## no missing values from count observation across all features
```

```python
appl_Data.drop(columns=['date','lights'],axis=1,inplace=True)
appl_Data.head()
```

```python
temp_data = appl_Data[['T2','T6']]

#(Take note of the initials of the defined variables below)

x_train, x_test,Y_train, Y_test = train_test_split(temp_data.drop(columns=['T6']),temp_data['T6'],test_size = 0.3,random_state = 42)
LinearRegression().fit(x_train,Y_train)
temp_predict = LinearRegression().fit(x_train,Y_train).predict(x_test)
temp_r2_score = r2_score(Y_test,temp_predict)
print(temp_r2_score)

print('R2 Score:{:.2f}'.format(temp_r2_score))
```

```python
## -already imported---from sklearn.preprocessing import MinMaxScaler

sca_appl_Data = MinMaxScaler().fit_transform(appl_Data)
sca_appl_Data = pd.DataFrame(sca_appl_Data ,columns = [appl_Data .columns])
sca_appl_Data 

X = sca_appl_Data.drop(columns=['Appliances'])
#X
y = sca_appl_Data[['Appliances']]
#Y


###Test and Train data
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size =0.3,random_state=42)
#X_train, X_test, y_train, y_test 

```

```python
appl_predict = LinearRegression().fit(X_train,y_train).predict(X_test)

mae = mean_absolute_error(y_test,appl_predict)
print('Mean Squared Error: {:.2f}'.format(mae))


rss = np.sum(np.square(y_test-appl_predict))
print('Residual Sum of Squares:',float(rss))

mse = np.sqrt(mean_squared_error(y_test,appl_predict))
print('Root mean squared error:{:.3f}'.format(mse))

r2_score = r2_score(y_test,appl_predict)
print('Coefficient of Determination:{:.2f}'.format(r2_score))

```

```python
from sklearn.linear_model import Lasso

l_regr = LinearRegression()
la_regr = Lasso(alpha = 0.001)

def feature_importance(model,feat):
    weight = model.fit(X_train,y_train).coef_
    weight= weight.reshape(26,1)
    feature_df = pd.DataFrame(weight,feat.columns).reset_index().sort_values(by=0)
    max = feature_df[feature_df[0] == feature_df[0].max()]
    min = feature_df[feature_df[0] == feature_df[0].min()]
    return(max,min,feature_df)
    
feature_importance(l_regr,X_train)
#feature_importance(la_regr,X_train)
```
